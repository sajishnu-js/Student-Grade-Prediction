```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Model Results
metrics = {
    'XGBoost' : [99.75, 1.00, 0.98, 0.99],
    'CatBoost': [99.79, 1.00, 0.99, 0.99],
    'PyTorch' : [99.74, 0.98, 1.00, 0.99],
}

#Creating Comparison Table
results = pd.DataFrame(metrics, index=['Accuracy', 'Precision', 'Recall', 'F1-Score'])
print(results)
```

```python
#Model Comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
colors  = ['blue', 'green', 'red', 'purple']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Model Comparison', fontsize=16,fontweight='bold')

for ax, metric,color in zip(axes.flat, metrics,colors):
    values = results.loc[metric]
    bars   = ax.bar(values.index, values.values,width=0.7, color=color)

    for bar, val in zip(bars, values.values):
        ax.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                str(val), ha='center', va='bottom')

    ax.set_title(metric)

plt.tight_layout()
plt.savefig('plots/model_comparison.png')
plt.show()
```
