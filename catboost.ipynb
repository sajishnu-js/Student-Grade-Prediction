```python
#Importing Libraries
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading Preprocessed Data
x_train = pd.read_csv('preprocessed_data/x_train.csv')
x_test  = pd.read_csv('preprocessed_data/x_test.csv')
y_train = pd.read_csv('preprocessed_data/y_train.csv').squeeze()
y_test  = pd.read_csv('preprocessed_data/y_test.csv').squeeze()
```

```python
#Building the CatBoost Model
cat_model = CatBoostClassifier(
    iterations    = 100,
    depth         = 6,
    learning_rate = 0.1,
    loss_function = 'MultiClass',
    random_seed   = 42,
    verbose       = 10
)
```

```python
cat_model.fit(x_train,y_train)
```

```python
y_pred_cat=cat_model.predict(x_test)
```

```python
#Accuracy score
print("Accuracy score for catboost model: ",round(accuracy_score(y_pred_cat,y_test)*100,2))
```

```python
#Classification report
print("classification report for catboost model")
print(classification_report(y_test,y_pred_cat))
```

```python
#CatBoost handled the minority class better than XGBoost.
```

```python
# Saving CatBoost Model
cat_model.save_model('saved_models/catboost_model.cbm')
print("CatBoost model saved!")
```
