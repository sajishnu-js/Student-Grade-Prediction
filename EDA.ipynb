```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x=df['grade'], color='red')
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('plots/plot1_grade_distribution.png')
plt.show()
```

```python
#Feature Distributions
feature=["weekly_self_study_hours","attendance_percentage","class_participation","total_score"]
df[features].hist(bins=50, figsize=(12, 8), color='green',edgecolor='white')
plt.suptitle('Feature Distributions')
plt.tight_layout()
plt.savefig('plots/plot2_feature_distributions.png')
plt.show()
```

```python
#Boxplots - Features vs Grade
f,a=plt.subplots(2,2,figsize=(15,10), dpi=100)
a=a.flatten()
for i,feature in enumerate(df.columns[1:5]):
    sns.boxplot(data=df, x=df["grade"],y=df[feature], ax=a[i])
    a[i].set_title(f"Boxplot of {feature} vs Grade")
plt.tight_layout()
plt.savefig('plots/plot3_boxplots.png')
plt.show()
```
