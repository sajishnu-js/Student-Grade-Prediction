```python
#Importing Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import torch
```

```python
#Loading the Dataset
df = pd.read_csv('dataset/student_performance.csv')
df.head()
```

```python
# Dropping the student_id Column
df = df.drop(["student_id"], axis=1)
df.head()
```

```python
# Encoding the Target Column - converting grades to numbers (A=0, B=1, C=2, D=3, F=4)
le = LabelEncoder()
df['grade'] = le.fit_transform(df['grade'])

print("Classes:", le.classes_)
print(df['grade'].value_counts())
```

```python
df.head()
```

```python
#Splitting Features and Target
x=df.drop(["grade"], axis=1)
y=df["grade"]
```

```python
#checking the features
x.head()
```

```python
#Checking the targets
y.head()
```

```python
#Splitting into Train and Test Sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42, stratify=y)

print("Train size :", x_train.shape)
print("Test size  :", x_test.shape)
```

```python
# Saving Preprocessed Data to CSV files
x_train.to_csv('preprocessed_data/x_train.csv', index=False)
x_test.to_csv('preprocessed_data/x_test.csv',   index=False)
y_train.to_csv('preprocessed_data/y_train.csv', index=False)
y_test.to_csv('preprocessed_data/y_test.csv',   index=False)
```

```python
#Scalling the features for PyTorch

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
# Converting Scaled Data to PyTorch Tensors
x_train_tensor=torch.tensor(x_train_scaled, dtype=torch.float32)
x_test_tensor=torch.tensor(x_test_scaled, dtype=torch.float32)

y_train_tensor=torch.tensor(y_train.values, dtype=torch.long)
y_test_tensor =torch.tensor(y_test.values,  dtype=torch.long)
```

```python
# Saving Tensors
torch.save(x_train_tensor, "preprocessed_data/x_train_tensor.pt")
torch.save(x_test_tensor,  "preprocessed_data/x_test_tensor.pt")
torch.save(y_train_tensor, "preprocessed_data/y_train_tensor.pt")
torch.save(y_test_tensor,  "preprocessed_data/y_test_tensor.pt")
```
